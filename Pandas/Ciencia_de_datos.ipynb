```python
import pandas as pd

# Lee el archivo .ipynb usando pandas
df = pd.read_json('Ciencia_de_datos.ipynb')

# Accede al contenido del cuaderno (notebook)
contenido = df['Estado_nomina.csv']

# Itera sobre las celdas y procesa el contenido
codigo_python = ''
for celda in contenido:
    if celda['cell_type'] == 'code':
        codigo_python += '\n'.join(celda['source']) + '\n\n'

# Corrige los caracteres especiales y la codificaci√≥n
# Puedes realizar otras operaciones de limpieza aqu√≠ si es necesario

# Guarda el c√≥digo procesado en un archivo .py
with open('archivo.py', 'w', encoding='utf-8') as f:
    f.write(codigo_python)
```
